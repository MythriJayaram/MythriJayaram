```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

```python
df_train = pd.read_csv("emnist-balanced-train.csv")
```

```python
df_train
```

```python
cols = df_train.columns.tolist()
#features = cols[-1:] + cols[:-1]
#df = df[cols]
target = df_train[cols[0:1]]
target

features = df_train[cols[1:]]
features.shape
```

```python
features_array = tf.keras.utils.normalize(features.to_numpy(), axis=1)


features_array

target_array = target.to_numpy()


#target_array = target.to_numpy()
#target_array
```

```python
features_array = np.resize(features_array, (112800, 28, 28))
```

```python
for x in range(len(features_array)):
    features_array[x] = features_array[x].transpose()
```

```python
df_test = pd.read_csv("emnist-mnist-test.csv")
```

cols = df_test.columns.tolist()
#features = cols[-1:] + cols[:-1]

target_test = df_test[cols[0:1]]

features_test = df_test[cols[1:]]

features_test = np.resize(features_test, (18800, 28, 28))

for x in range(len(features_test)):
    features_test[x] = features_test[x].transpose()

features_test_array = tf.keras.utils.normalize(features_test, axis=1)

target_test_array = target_test.to_numpy()

target_test.shape

features_test_array.shape

```python
features_test_array.shape


target_test_array.shape
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv2D(28, (3, 3), activation="relu", input_shape=(28, 28, 1)));

model.add(tf.keras.layers.MaxPooling2D((2, 2)));

model.add(tf.keras.layers.Conv2D(56, (3, 3), activation="relu"));

model.add(tf.keras.layers.MaxPooling2D((2, 2)));
  
model.add(tf.keras.layers.Flatten());

model.add(tf.keras.layers.Dense(units=47, activation=tf.nn.softmax))

# Compiling and optimizing model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training the model
model.fit(features_array, target_array, epochs=10)

# Saving the model
model.save('handwritten_digits.model')
```

```python
val_loss, val_acc = model.evaluate(features_test_array, target_test_array)
print(val_loss)
print(val_acc)
```
